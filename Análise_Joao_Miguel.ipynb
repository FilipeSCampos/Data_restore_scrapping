```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/product_prices.csv')
data
```

```python
data.info()
```

```python
# Criar DataFrame
df = pd.DataFrame(data)

# Extrair preÃ§os
df['Price_1'] = df['Price'].str.extract(r'R\$(\d+\.?\d*)')[0]
df['Price_2'] = df['Price'].str.extract(r'em (\d+\.?\d*)')[0]

# Exibir DataFrame resultante
print(df)
```

```python
# Remover as colunas 'Price_1' e 'Price_2'
df.drop(columns=['Price_1', 'Price_2'], inplace=True)
```

```python
# Remover todos os valores NaN do DataFrame
df = df.dropna()
```

```python
df
```

```python
# Converter a coluna 'Date' para o tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Converter a coluna 'Price' para o tipo float
df['Price'] = df['Price'].astype(float)

# Exibir o DataFrame resultante
print(df)
```

```python
df.info()
```

```python
# Plotar o grÃ¡fico de histÃ³rico de preÃ§os
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price'], marker='o', linestyle='-')
plt.title('HistÃ³rico de PreÃ§os')
plt.xlabel('Data')
plt.ylabel('PreÃ§o')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
